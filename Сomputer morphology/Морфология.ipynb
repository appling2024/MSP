OPEN FILE

```python
with open('brusov.txt', "r", encoding="utf-8") as file:
    text=file.read()
```

SPACY

```python
import spacy
nlp = spacy.load("ru_core_news_sm")
```

```python
doc = nlp(text)
import pandas as pd

cols = ("text", "lemma", "POS", "explain", "tense")
rows = []

for t in doc:
    row = [t.text, t.lemma_, t.pos_, spacy.explain(t.pos_), t.morph]
    rows.append(row)

df = pd.DataFrame(rows, columns=cols)
    
df
```

PYMORPHY2

```python
import pymorphy3
from pymorphy3.tokenizers import simple_word_tokenize
```

```python
morph = pymorphy3.MorphAnalyzer()
```

```python
tokens = simple_word_tokenize(text)
cols = ("word", "normal_form", "tag")
rows = []
for i in tokens:
    parsed=morph.parse(i)
    if parsed:
        p = parsed[0]
        row=[i, p.normal_form, p.tag]
        rows.append(row)
df = pd.DataFrame(rows, columns=cols)
print(df)
```

```python
import json
import pandas as pd
from pymystem3 import Mystem

m = Mystem()
analysis_result = m.analyze(text)

data = []

for item in analysis_result:
    if 'analysis' in item:
        for analysis in item['analysis']:
            data.append({
                'text': item['text'],
                'lex': analysis['lex'],
                'weight': analysis['wt'],
                'grammar': analysis['gr']
            })
    else:
        data.append({
            'text': item['text'],
            'lex': None,
            'weight': None,
            'grammar': None
        })

df = pd.DataFrame(data)

print(df)
```
