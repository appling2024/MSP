```python
#nltk
with open('/content/статья.txt' , 'r' , encoding='utf-8') as f:
  my_file = f.read()

tokens = nltk.word_tokenize(my_file)
print(tokens)
print(len(tokens))

with open('tokens_nltk.txt' , 'w') as f2:
  for item in tokens:
    f2.write(item)
    f2.write('\n')
```

```python
#stanza
with open('/content/статья.txt' , 'r') as f:
  my_file = f.read()

ppln = stanza.Pipeline('ru', processors='tokenize')
doc = ppln(my_file)
tokens = 0
for i, sentence in enumerate(doc.sentences):
    #print(f'====== Предложение {i+1} =======')
    #print(*[f'id: {token.id}\ttext: {token.text}' for token in sentence.tokens], sep='\n')
    for token in sentence.tokens:
      #print(token.text)
      tokens = tokens + 1

print('\nВсего токенов: ' , tokens)
print('\nВсего токенов коротким способом: ' , doc.num_tokens)

with open('tokens_stanza.txt' , 'w') as f2:
  for i, sentence in enumerate(doc.sentences):
    for token in sentence.tokens:
      f2.write(token.text)
      f2.write('\n')
```

```python
# UDPipe
from ufal.udpipe import Model, Pipeline

with open('/content/статья.txt', 'r', encoding='utf-8') as f:
    my_file = f.read()

model = Model.load('russian-ud-2.0-170801.udpipe')
pipeline = Pipeline(model, 'tokenize', Pipeline.DEFAULT, Pipeline.DEFAULT, Pipeline.DEFAULT)

tokens = pipeline.process(my_file)

tokenized_ud = []
for sent in tokens.split('\n\n'):
    for token in sent.split('\n'):
        if token and not token.startswith('#'):
            tokenized_ud.append(token.split('\t')[1])

with open('tokens_UDPipe.txt', 'w', encoding='utf-8') as f2:
    for token in tokenized_ud:
        f2.write(token + '\n')
```
