```python
#nltk

with open('статья.txt' , 'r' , encoding='utf-8') as f:
  my_file = f.read()

import nltk
from nltk.tokenize import word_tokenize
tokenized_nltk = word_tokenize(my_file, language='russian')

tokens = word_tokenize(my_file)
print(tokens)
print(len(tokens))

with open('tokens_nltk.txt' , 'w', encoding='utf-8') as f2:
  for item in tokens:
    f2.write(item)
    f2.write('\n')
```

```python
#stanza
with open('статья.txt' , 'r', encoding='utf-8') as f:
  my_file = f.read()

import stanza
ppln = stanza.Pipeline('ru', processors='tokenize')
doc = ppln(my_file)
tokens = 0
for i, sentence in enumerate(doc.sentences):
    #print(f'====== Предложение {i+1} =======')
    #print(*[f'id: {token.id}\ttext: {token.text}' for token in sentence.tokens], sep='\n')
    for token in sentence.tokens:
      #print(token.text)
      tokens = tokens + 1

print('\nВсего токенов: ' , tokens)
print('\nВсего токенов коротким способом: ' , doc.num_tokens)

with open('tokens_stanza.txt' , 'w', encoding='utf-8') as f2:
  for i, sentence in enumerate(doc.sentences):
    for token in sentence.tokens:
      f2.write(token.text)
      f2.write('\n')
```

```python
# UDPipe
import ufal.udpipe
from ufal.udpipe import Model, Pipeline

with open('статья.txt', 'r', encoding='utf-8') as f:
    my_file = f.read()

model = Model.load('russian-ud-2.0-170801.udpipe')
pipeline = Pipeline(model, 'tokenize', Pipeline.DEFAULT, Pipeline.DEFAULT, Pipeline.DEFAULT)

tokens = pipeline.process(my_file)

tokenized_ud = []
for sent in tokens.split('\n\n'):
    for token in sent.split('\n'):
        if token and not token.startswith('#'):
            tokenized_ud.append(token.split('\t')[1])

with open('tokens_UDPipe.txt', 'w', encoding='utf-8') as f2:
    for token in tokenized_ud:
        f2.write(token + '\n')
```

```python
import difflib

def read_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.readlines()

def compare_tokenizations_in_columns(file1, file2, column_width=30):
    tokens1 = read_file(file1)  
    tokens2 = read_file(file2) 

    max_len = max(len(tokens1), len(tokens2))

    tokens1 += [''] * (max_len - len(tokens1))
    tokens2 += [''] * (max_len - len(tokens2))

    print(f"{file1:<{column_width}} | {file2:<{column_width}}")
    print("-" * (column_width * 2 + 3))

    for tok1, tok2 in zip(tokens1, tokens2):
        print(f"{tok1.strip():<{column_width}} | {tok2.strip():<{column_width}}")

file_path1 = "tokens_stanza.txt"
file_path2 = "tokens_spaCy.txt"
file_path3 = "tokens_razdel.txt"
file_path4 = "tokens_nltk.txt"
file_path5 = "tokens_pymorphy.txt"
file_path6 = "tokens_UDPipe.txt"
file_path7 = "tokens_segtok.txt"

def compare_multiple_files_in_columns(*file_paths, column_width=30):
    all_tokens = [read_file(file_path) for file_path in file_paths]

    max_len = max(len(tokens) for tokens in all_tokens)

    for tokens in all_tokens:
        tokens += [''] * (max_len - len(tokens))

    header = "\t".join(f"{file_path.split('/')[-1]:<{column_width}}" for file_path in file_paths)
    print(header)
    print("-" * (column_width * len(file_paths) + len(file_paths) - 1))

    for row in zip(*all_tokens):
        print("\t".join(f"{token.strip():<{column_width}}" for token in row))

compare_multiple_files_in_columns(
    file_path1, file_path2, file_path3, file_path4, file_path5, file_path6, file_path7, column_width=20
)
```
