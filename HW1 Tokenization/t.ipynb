NLTK

```python
import nltk
with open ('war_piece_Ntl (1).txt', 'r', encoding='utf-8') as f:
    my_file = f.read()

import nltk
from nltk.tokenize import word_tokenize
tokenized_nltk = word_tokenize(my_file, language='russian')

tokens = word_tokenize(my_file)
print(tokens)
print(len(tokens))

with open('tokens_nltk.txt' , 'w', encoding='utf-8') as f2:
    for item in tokens:
        f2.write(item)
        f2.write('\n')
```

Stanza

```python
import stanza
with open('war_piece_Ntl(1).txt', 'r', encoding='utf-8') as f:
  my_file = f.read()

ppln = stanza.Pipeline('ru', processors='tokenize') 

doc = ppln(my_file)

print('\nВсего токенов коротким способом:', doc.num_tokens)

with open('tokens_stanza.txt', 'w', encoding='utf-8') as f2:
  for i, sentence in enumerate(doc.sentences):
    for token in sentence.tokens:
      f2.write(token.text)
      f2.write('\n')
```

Razdel


```python
from razdel import tokenize
tokenized_razdel = [tk.text for tk in tokenize(my_file)]
print(tokenized_razdel)
print(len(tokenized_razdel))

with open('tokens_razdel.txt' , 'w', encoding='utf-8') as f2:
    for token in tokenized_razdel:
        f2.write(token)
        f2.write('\n')
        
```

Segtok

```python
from segtok.tokenizer import word_tokenizer
tokenized_segtok=word_tokenizer(my_file)
print(tokenized_segtok)
print(len(tokenized_segtok))

with open('tokens_segtok.txt' , 'w', encoding='utf-8') as f2:
    for item in tokenized_segtok:
        f2.write(item)
        f2.write('\n')
```

Spacy

```python
import spacy
nlp = spacy.load('ru_core_news_sm')
tokenized_spacy = [tk.text for tk in nlp(my_file) if tk.text != ' ']

print(tokenized_spacy)
print(len(tokenized_spacy))

with open('tokens_spacy.txt' , 'w', encoding='utf-8') as f2:
    for item in tokenized_spacy:
        f2.write(item)
        f2.write('\n')
```

```python
import  pymorphy3
with open ('war_piece_Ntl (1).txt', 'r', encoding='utf-8') as f:
    my_file = f.read()
    
    from pymorphy3.tokenizers import simple_word_tokenize
tokenized_pymorphy3 = simple_word_tokenize(my_file)

tokens = word_tokenize(my_file)
print(tokenized_pymorphy3)
print(len(tokenized_pymorphy3))

with open('tokens_pymorphy3.txt' , 'w', encoding='utf-8') as f2:
    for item in tokenized_pymorphy3:
        f2.write(item)
        f2.write('\n')
```

UdPipe

```python
import ufal.udpipe
from ufal.udpipe import Model, Pipeline

with open('war_piece_Ntl(1).txt', 'r', encoding='utf-8') as f:
    my_file = f.read()

model = Model.load('russian-ud-2.0-170801.udpipe')
pipeline = Pipeline(model, 'tokenize', Pipeline.DEFAULT, Pipeline.DEFAULT, Pipeline.DEFAULT)

tokens = pipeline.process(my_file)

tokenized_ud = []
for sent in tokens.split('\n\n'):
    for token in sent.split('\n'):
        if token and not token.startswith('#'):
            tokenized_ud.append(token.split('\t')[1])

with open('tokens_UDPipe.txt', 'w', encoding='utf-8') as f2:
      for item in tokenized_ud:
        f2.write(item)
        f2.write('\n')  
        
print(len(tokenized_ud))
```
