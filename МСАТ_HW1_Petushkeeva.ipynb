#### NLTK

```python
! pip install nltk
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
with open('север 1918.txt', 'r', encoding='utf-8') as f:
	zam = f.read().replace('\n', ' ')
```

```python
len(zam), len(zam.split())
```

```python
from nltk import word_tokenize
zam_nltk = nltk.word_tokenize(zam, language = 'russian')
```

```python
zam_nltk
```

```python
nlp = stanza.Pipeline(lang='ru', processors='tokenize', verbose=False)
zam_stanza = [tk.text for tk in nlp(zam).iter_tokens()]
```

#### Сравнение

```python
lengths = [len(zam_nltk), len(zam_razdel), len(zam_segtok), len(zam_spacy), len(zam_pymorphy3), len(zam_moses), len(zam_stanza)]
lengths
```

```python
m = max(lengths)
print(m)
print([m-l for l in lengths])
```

```python
zam_nltk += ['' for i in range(m-len(zam_nltk))]
zam_razdel += [''  for i in range(m-len(zam_razdel))]
zam_segtok += [''  for i in range(m-len(zam_segtok))]
zam_spacy += [''  for i in range(m-len(zam_spacy))]
zam_pymorphy3 += [''  for i in range( m-len(zam_pymorphy3))]
zam_moses += [''  for i in range( m-len(zam_moses))]
zam_stanza += [''  for i in range( m-len(zam_stanza))]
```

```python
print('{:20}\t{:20}\t{:20}\t{:20}\t{:20}\t{:20}\t{:20}'.format('nltk', 'razdel', 'segtok', 'spacy', 'pymorphy3', 'moses', 'stanza'))
for tk in zip(zam_nltk, zam_razdel, zam_segtok,zam_spacy, zam_pymorphy3, zam_moses, zam_stanza):
    print('{:20}\t{:20}\t{:20}\t{:20}\t{:20}\t{:20}\t{:20}'.format(*tk))
```
