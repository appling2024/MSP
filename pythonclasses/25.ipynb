```python
with open('salinger1.txt', 'r', encoding='utf-8') as f:
    my_file = f.read().lower()

my_file = my_file.replace('"','')
my_file = my_file.replace('-','')
my_file = my_file.replace('.','')
my_file = my_file.replace(';','')
my_file = my_file.replace(',','')
my_file = my_file.replace('!','')
my_file = my_file.replace('?','')
my_file = my_file.replace(':','')

words_raw = my_file.split()
words = []
for word in words_raw:
    if word not in words:
        words.append(word)
words.sort()

print(words)
print(len(words))
```

```python
with open('salinger1.txt', 'r', encoding='utf-8') as f:
    my_file = f.read().lower()

my_file = my_file.replace('"','')
my_file = my_file.replace('-','')
my_file = my_file.replace('.','')
my_file = my_file.replace(';','')
my_file = my_file.replace(',','')
my_file = my_file.replace('!','')
my_file = my_file.replace('?','')
my_file = my_file.replace(':','')

import nltk
from nltk.tokenize import word_tokenize
tokenized_nltk = word_tokenize(my_file, language='russian')

words = []
for word in tokenized_nltk:
   if word not in words:
      words.append(word)
words.sort()

print(words)
print(len(words))
```
