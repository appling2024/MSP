```python
from __future__ import unicode_literals
import re
```

```python
GROUPING_SPACE_REGEX = re.compile(r'([^\w_-]|[+])', re.UNICODE)
```

```python
def simple_word_tokenize(text, _split=GROUPING_SPACE_REGEX.split):
    """
    Split text into tokens. Don't split by a hyphen.
    Preserve punctuation, but not whitespaces.
    """
    return [t for t in _split(text) if t and not t.isspace()]
```

```python
from ufal.udpipe import Model, Pipeline
```

```python
model = Model.load('russian-ud-2.0-170801.udpipe')
```

```python
pipeline = Pipeline(model, 'tokenize', Pipeline.DEFAULT, Pipeline.DEFAULT, Pipeline.DEFAULT)
```

```python
tokens = pipeline.process(text)
tokenized_ud = []
for sentence in tokens.split('\n\n'):
    for token in sentence.split('\n'):
        if (not token.startswith('#')) and token != '' :
            tokenized_ud.append(token.split('\t')[1])
```
